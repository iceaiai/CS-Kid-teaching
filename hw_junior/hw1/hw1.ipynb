# Operators

```python
import random
import ipywidgets as widgets
from IPython.display import display, clear_output

# Function to evaluate the expression
def evaluate_expression(expression):
    try:
        return eval(expression)
    except Exception as e:
        return f"Error: {e}"

# Function to generate random presets
def generate_presets(num_problems=50, max_value=200):
    presets = []
    for _ in range(num_problems):
        numbers = [random.randint(1, max_value) for _ in range(3)]
        operators = ['+', '-', '*', '/', '**']

        # Randomly choose operators for the correct expression
        op1 = random.choice(operators)
        op2 = random.choice(operators)

        # Evaluate to find the target
        expression = f"{numbers[0]} {op1} {numbers[1]} {op2} {numbers[2]}"
        target = evaluate_expression(expression)

        # Store preset
        presets.append({
            'numbers': numbers,
            'target': target,
            'correct_op1': op1,
            'correct_op2': op2
        })
    return presets

# Generate 50 presets
presets = generate_presets()

# Choose a random preset for the game
current_preset = random.choice(presets)

# Create input widgets
numbers_display = widgets.HTML(value=f"<b>Numbers:</b> {current_preset['numbers']}")
target_display = widgets.HTML(value=f"<b>Target:</b> {current_preset['target']}")
op1_input = widgets.Text(description="Operator 1", placeholder="+, -, *, /, **")
op2_input = widgets.Text(description="Operator 2", placeholder="+, -, *, /, **")
check_button = widgets.Button(description="Check Expression")
result_display = widgets.HTML(value="")

# Function to check the student's answer
def check_expression(b):
    clear_output(wait=True)
    display(numbers_display, target_display, op1_input, op2_input, check_button, result_display)

    first_operator = op1_input.value
    second_operator = op2_input.value
    numbers = current_preset['numbers']
    target = current_preset['target']

    # Create the student's expression
    expression = f"{numbers[0]} {first_operator} {numbers[1]} {second_operator} {numbers[2]}"
    result = evaluate_expression(expression)

    # Check the result
    if result == target:
        result_display.value = f"<b>Congratulations!</b> The expression <code>{expression}</code> equals the target {target}."
    else:
        correct_expression = f"{numbers[0]} {current_preset['correct_op1']} {numbers[1]} {current_preset['correct_op2']} {numbers[2]}"
        result_display.value = (f"<b>Try Again!</b> The expression <code>{expression}</code> equals {result}, not {target}. "
                                f"The correct expression was <code>{correct_expression}</code>.")

# Display widgets
display(numbers_display, target_display, op1_input, op2_input, check_button, result_display)

# Assign the button click event to the check_expression function
check_button.on_click(check_expression)
```

# Variables

```python
import random
import ipywidgets as widgets
from IPython.display import display, clear_output

# Initialize the score
score = 0

# Function to check the student's answer
def check_answer(question_num, correct_answer, student_answer):
    global score
    if str(student_answer) == str(correct_answer):
        score += 1
        return f"Correct! Your score is now {score}."
    else:
        return f"Incorrect. The correct answer was {correct_answer}. Your score is {score}."

# Integer Questions
def generate_integer_presets():
    presets = []
    for _ in range(20):
        a = random.randint(1, 50)
        b = random.randint(1, 50)
        c = random.randint(1, 50)
        expression = f"{a} + {b} * {c}"
        answer = eval(expression)
        presets.append((expression, answer))
    return presets

# Float Questions
def generate_float_presets():
    presets = []
    for _ in range(20):
        a = round(random.uniform(1, 50), 2)
        b = round(random.uniform(1, 50), 2)
        c = round(random.uniform(1, 50), 2)
        expression = f"{a} * {b} + {c}"
        answer = eval(expression)
        presets.append((expression, answer))
    return presets

# Boolean Questions
def generate_boolean_presets():
    presets = []
    for _ in range(20):
        a = random.randint(1, 50)
        b = random.randint(1, 50)
        c = random.randint(1, 50)
        expression = f"{a} > {b} and {b} < {c}"
        answer = eval(expression)
        presets.append((expression, answer))
    return presets

# String Questions
def generate_string_presets():
    presets = []
    for _ in range(20):
        word1 = random.choice(['Hello', 'World', 'Python', 'Coding'])
        word2 = random.choice(['Fun', 'Great', 'Awesome', 'Challenging'])
        expression = f"'{word1}' + ' ' + '{word2}' + '!'"
        answer = eval(expression)
        presets.append((expression, answer))
    return presets

# Generate all presets
integer_presets = generate_integer_presets()
float_presets = generate_float_presets()
boolean_presets = generate_boolean_presets()
string_presets = generate_string_presets()

# Function to create integer question
def integer_question():
    global score
    clear_output(wait=True)
    question = random.choice(integer_presets)
    display(widgets.HTML(value=f"<b>Integer Question:</b> What is the result of {question[0]}?"))
    student_answer = widgets.IntText(description="Answer:")
    check_button = widgets.Button(description="Check Answer")

    def check(b):
        result_display.value = check_answer(1, question[1], student_answer.value)

    result_display = widgets.HTML(value="")
    check_button.on_click(check)
    display(student_answer, check_button, result_display)

# Function to create float question
def float_question():
    global score
    clear_output(wait=True)
    question = random.choice(float_presets)
    display(widgets.HTML(value=f"<b>Float Question:</b> What is the value of {question[0]}?"))
    student_answer = widgets.FloatText(description="Answer:")
    check_button = widgets.Button(description="Check Answer")

    def check(b):
        result_display.value = check_answer(2, question[1], student_answer.value)

    result_display = widgets.HTML(value="")
    check_button.on_click(check)
    display(student_answer, check_button, result_display)

# Function to create boolean question
def boolean_question():
    global score
    clear_output(wait=True)
    question = random.choice(boolean_presets)
    display(widgets.HTML(value=f"<b>Boolean Question:</b> Is {question[0]}? (True/False)"))
    student_answer = widgets.Dropdown(options=["True", "False"], description="Answer:")
    check_button = widgets.Button(description="Check Answer")

    def check(b):
        result_display.value = check_answer(3, question[1], student_answer.value == "True")

    result_display = widgets.HTML(value="")
    check_button.on_click(check)
    display(student_answer, check_button, result_display)

# Function to create string question
def string_question():
    global score
    clear_output(wait=True)
    question = random.choice(string_presets)
    display(widgets.HTML(value=f"<b>String Question:</b> What is the result of {question[0]}?"))
    student_answer = widgets.Text(description="Answer:")
    check_button = widgets.Button(description="Check Answer")

    def check(b):
        result_display.value = check_answer(4, question[1], student_answer.value)

    result_display = widgets.HTML(value="")
    check_button.on_click(check)
    display(student_answer, check_button, result_display)

# Function to create the main menu
def main_menu():
    clear_output(wait=True)
    display(widgets.HTML(value="<h2>Welcome to the Variable Practice Game!</h2>"))

    int_button = widgets.Button(description="Integer Question")
    float_button = widgets.Button(description="Float Question")
    bool_button = widgets.Button(description="Boolean Question")
    string_button = widgets.Button(description="String Question")

    display(int_button, float_button, bool_button, string_button)

    int_button.on_click(lambda b: integer_question())
    float_button.on_click(lambda b: float_question())
    bool_button.on_click(lambda b: boolean_question())
    string_button.on_click(lambda b: string_question())

# Start the game with the main menu
main_menu()
```
